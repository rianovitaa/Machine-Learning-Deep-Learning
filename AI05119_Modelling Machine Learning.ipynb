## Import Library

```python
import pandas as pd
import numpy as np
```

## Input Dataset

```python
#Membaca data dan memasukkannya ke dalam bentuk Pandas Dataframe
df = pd.read_csv('https://raw.githubusercontent.com/Rietaros/kampus_merdeka/main/SC_HW1_bank_data.csv')
```

```python
#Jalankan code untuk mengecek nama kolom yang tersedia
df.columns
```

```python
df.head()
```

```python
#Hilangkan kolom yang dirasa tidak relevan dengan model (contoh: ID). None dapat diisi dengan nama-nama kolom yang akan digunakan.
#Contoh df = df[['X1','X2', 'Y']].copy()

#START CODE
df = df[['Geography', 'Gender', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary', 'Exited']].copy()
df
```

## Preprocessing

```python
#Lakukan One-Hot Encoder untuk data categorical, dengan fungsi pandas get_dummies

#START CODE
df = pd.get_dummies(df, columns=['Geography', 'Gender'], drop_first=True)

df
```

```python
#Pisahkan mana X (feature) dengan Y,
#Y adalah kolom "Exited"

#START CODE
X = df.drop(columns=['Exited'])  # Drop the 'Exited' column to get X
y = df['Exited']  # Extract the 'Exited' column as Y
```

```python
# Display X
print("Features (X):")
X.head()
```

```python
# Display y
print("Target Variable (Y):")
y.head()
```

```python
#Lakukan Scaler dan/atau Noermalisasi jika diperlukan
from sklearn.preprocessing import MinMaxScaler

# Initialize the Min-Max scaler
scaler = MinMaxScaler()

# Fit and transform the scaler on your feature data (X)
X_transformed = scaler.fit_transform(X)
```

```python
#Ini digunakan jika dilakukan scaler/Normalisas. Jika tidak, code ini bisa dilewat dan diganti dengan code yang ada di dalam komen
X_transform = pd.DataFrame(X_transformed, columns = X.columns)
#X_transform = X.copy()
```

## Train-Test Split

```python
#Split menjadi train dan test dengan test_size 25%
#Tidak perlu mengubah code ini

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_transformed,y,test_size = 0.25,random_state = 123)
```

# Modeling

## Model1

Jelaskan secara Singkat Model ke-1 yang digunakan!

**LOGISTIC REGRESSION**

**Logistic Regression digunakan untuk memprediksi probabilitas sukses atau kegagalan suatu peristiwa berdasarkan satu atau beberapa variabel input.
Output dari model Logistic Regression adalah probabilitas dalam rentang 0 hingga 1, yang dapat diinterpretasikan sebagai kemungkinan kejadian tertentu. Dalam klasifikasi biner, model ini membagi data menjadi dua kelas (biasanya "1" dan "0") berdasarkan probabilitas prediksi. Untuk menghasilkan prediksi, Logistic Regression menggunakan fungsi logistik (sigmoid) yang melengkung yang mengubah hasil dari kombinasi linear fitur menjadi probabilitas. Namun, logistic regression memiliki batasan ketika menangani masalah dengan kompleksitas tinggi atau ketika ada hubungan non-linear antara fitur dan target**

```python
#Pilih salah satu metode Machine Leaarning
#Model Machine Learning dapat dipanggil terlebih dahulu melalui library yang digunakan. Gunakan library scikit learn seperti pada contoh

# Import the machine learning model you want to use, for example, LogisticRegression
from sklearn.linear_model import LogisticRegression

# Initialize the model
model1 = LogisticRegression()

# Define the hyperparameters and their possible values
params = {
    "tol": [0.1, 0.01, 0.001],
    'C': [0.5, 1.0, 1.5, 2.0]
}

# Import necessary libraries for hyperparameter tuning
from sklearn.model_selection import GridSearchCV

# Create a GridSearchCV object
grid = GridSearchCV(
    estimator=model1,
    param_grid=params,
    scoring='accuracy',
    n_jobs=10,  # Number of CPU cores to use
    cv=10  # Number of cross-validation folds
)

# Fit the model on your training data
grid.fit(X_train, y_train)

# Get the best hyperparameters from the grid search
best_params = grid.best_params_
print("Best Hyperparameters:", best_params)
```

```python
#lakukan evaluasi dengan beberapa maetric di bawah ini
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred = grid.predict(X_test)

print(classification_report(y_test,y_pred))
print("")
print(confusion_matrix(y_test,y_pred))
print("")
print(accuracy_score(y_test, y_pred))
```

**DECISION TREE**


**Decision Tree adalah model yang mengambil bentuk pohon, dengan node akar sebagai awalnya dan node-node daun sebagai hasil akhirnya. Hasil pemilihan fitur tersebut mengarahkan model ke node anak yang sesuai, dan proses ini berlanjut hingga mencapai node daun, di mana keputusan atau prediksi akhir dibuat. Decision Tree mudah dipahami dan dapat diinterpretasikan, cocok untuk data dengan fitur kategorikal atau numerik, dan digunakan baik untuk masalah klasifikasi maupun regresi. Namun, mereka dapat cenderung overfitting jika tidak diatur dengan baik dan sensitif terhadap perubahan kecil dalam data pelatihan.**



```python
from sklearn.tree import DecisionTreeClassifier

# Initialize the Decision Tree model
model2 = DecisionTreeClassifier()

# Define the hyperparameters for the Decision Tree Classifier
params = {
    "criterion": ["gini", "entropy"],
    "max_depth": [None, 10, 20, 30],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}

# Import library yang diperlukan untuk melakukan GridSearchCV
from sklearn.model_selection import GridSearchCV

# Buat objek GridSearchCV
grid2 = GridSearchCV(
    estimator=model2,
    param_grid=params,
    scoring='accuracy',
    n_jobs=10,  # Jumlah core CPU yang digunakan
    cv=10  # Jumlah lipatan cross-validation
)

# Lakukan penyetelan parameter pada data pelatihan
grid2.fit(X_train, y_train)

# Dapatkan hyperparameter terbaik dari hasil penyetelan
best_params = grid2.best_params_
print("Best Hyperparameters:", best_params)
```

```python
#lakukan evaluasi dengan beberapa maetric di bawah ini
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred = grid2.predict(X_test)

print(classification_report(y_test,y_pred))
print("")
print(confusion_matrix(y_test,y_pred))
print("")
print(accuracy_score(y_test, y_pred))
```

 **GRADIENTBOOSTING CLASSIFIER**
 **GradientBoostingClassifier adalah bagian dari keluarga algoritma "Boosting," yang berarti menggabungkan beberapa model lemah menjadi satu model kuat. Algoritma ini bekerja dengan cara secara berurutan membangun serangkaian Decision Trees kecil yang disebut "weak learners" dan memperbaiki kesalahan prediksi sebelumnya pada setiap langkah. Gabungan dari serangkaian Decision Trees ini membentuk model yang kuat yang dapat digunakan untuk klasifikasi. GradientBoostingClassifier menggunakan teknik gradien (gradient descent) untuk mengoptimalkan modelnya dengan mengurangi kesalahan prediksi secara berangsur-angsur. Namun, perlu diatur dengan hati-hati untuk menghindari overfitting, dan dapat memerlukan waktu yang lebih lama untuk melatih dibandingkan dengan model lainnya.**

```python
from sklearn.ensemble import GradientBoostingClassifier

# Initialize the Gradient Boosting Classifier
model = GradientBoostingClassifier(n_estimators=300, max_depth=3, random_state=42)

# Fit the model on the training data
model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = model.predict(X_test)
```

```python
#lakukan evaluasi
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred = model.predict(X_test)

print(classification_report(y_test,y_pred))
print("")
print(confusion_matrix(y_test,y_pred))
print("")
print(accuracy_score(y_test, y_pred))
```

**KNEIGHBORS CLASSIFIER**

**KNeighborsClassifier adalah algoritma berbasis instansi (instance-based) yang digunakan untuk masalah klasifikasi. Algoritma ini bekerja dengan cara "mengenal tetangga" dari setiap data pelatihan dan memutuskan kelas dari data uji berdasarkan mayoritas kelas tetangga terdekatnya. Kelemahan KNeighborsClassifier adalah kinerjanya yang lambat jika datasetnya besar, karena harus memeriksa jarak ke setiap titik data pelatihan. Algoritma ini juga sensitif terhadap skala data, sehingga preprocessing seperti penskalaan fitur seringkali diperlukan.**

```python
from sklearn.neighbors import KNeighborsClassifier

# Initialize the K-Nearest Neighbors (KNN) model
model4 = KNeighborsClassifier()

# Define the hyperparameters for the KNN Classifier
params = {
    "n_neighbors": [5, 15, 20, 5],
    "weights": ["uniform", "distance"],
    "algorithm": ["auto", "ball_tree", "kd_tree", "brute"]
}

# Import library yang diperlukan untuk melakukan GridSearchCV
from sklearn.model_selection import GridSearchCV

# Buat objek GridSearchCV
grid4 = GridSearchCV(
    estimator=model4,
    param_grid=params,
    scoring='accuracy',
    n_jobs=10,  # Jumlah core CPU yang digunakan
    cv=10  # Jumlah lipatan cross-validation
)

# Lakukan penyetelan parameter pada data pelatihan
grid4.fit(X_train, y_train)

# Dapatkan hyperparameter terbaik dari hasil penyetelan
best_params = grid4.best_params_
print("Best Hyperparameters:", best_params)
```

```python
#lakukan evaluasi dengan beberapa maetric di bawah ini
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred = grid4.predict(X_test)

print(classification_report(y_test,y_pred))
print("")
print(confusion_matrix(y_test,y_pred))
print("")
print(accuracy_score(y_test, y_pred))
```

**RANDOM FOREST CLASSIFIER**

**RandomForestClassifier adalah algoritma ensemble learning yang membangun sejumlah besar Decision Trees (pohon keputusan) dan menggabungkan hasil prediksi dari pohon-pohon tersebut. Setiap pohon dalam Random Forest dibangun dengan menggunakan subset acak dari data pelatihan (bootstrap) dan subset acak dari fitur (subsetting). Algoritma ini efektif dalam mengatasi overfitting karena penggunaan banyak pohon yang beragam. Algoritma ini cocok untuk berbagai jenis data dan masalah klasifikasi, termasuk data dengan fitur kategorikal atau numerik. Meskipun Random Forest kuat, pengguna harus memperhatikan beberapa hiperparameter seperti jumlah pohon dan kedalaman pohon untuk mengoptimalkan performa model.**

```python
from sklearn.ensemble import RandomForestClassifier
# Initialize the Random Forest Classifier
model = RandomForestClassifier(n_estimators=100, max_depth=20, random_state=42)

# Fit the model on the training data
model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = model.predict(X_test)
```

```python
#lakukan evaluasi dengan beberapa maetric di bawah ini
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred = model.predict(X_test)

print(classification_report(y_test,y_pred))
print("")
print(confusion_matrix(y_test,y_pred))
print("")
print(accuracy_score(y_test, y_pred))
```
